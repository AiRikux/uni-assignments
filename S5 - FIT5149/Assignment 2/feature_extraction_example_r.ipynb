```r
df = read.csv('./train_data_withlabels.csv') 
```

```r
window_left = 50
window_right = 50
window_size = window_left + window_right +1
```

```r
# During to window size, the leftest and rightest 50 points(100 in total) could not have window-based features
add_remnants <- function(temp_l, window_left, window_right){
    temp_l <- append(rep(0, window_left), temp_l)
    temp_l <- append(temp_l, rep(0, window_right))
}
```

```r
# Using re_diff/ab_diff as example in each window (right_value - left_value)
df$new_rediff <- add_remnants(diff(df$total_load, lag=(window_left + window_right)), window_left, window_right)
df$new_abdiff <- add_remnants(abs(diff(df$total_load, lag=(window_left + window_right))), window_left, window_right)
```

```r
# Calculate mean
df$new_mean <- add_remnants(rollmean(df$total_load,window_size), window_left, window_right)

# Calculate std
df$new_std <- add_remnants(rollapplyr(df$total_load,window_size, sd), window_left, window_right)

# Calculate peak
df$new_peak <- add_remnants(rollmax(df$total_load,window_size), window_left, window_right)

# Calculate peak
df$new_peak <- add_remnants(rollmax(df$total_load,window_size), window_left, window_right)
```

```r
num_crossing_points <- function(x) {
  midline <- median(x, na.rm = TRUE)
  ab <- x <= midline
  lenx <- length(x)
  p1 <- ab[1:(lenx - 1)]
  p2 <- ab[2:lenx]
  cross <- (p1 & !p2) | (p2 & !p1)
  return(sum(cross, na.rm = TRUE))
}
```

```r
# crossing point
df$new_cp <- add_remnants(rollapply(df$total_load, FUN=num_crossing_points, width = window_size), window_left, window_right)
```
